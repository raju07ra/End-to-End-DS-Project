```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 7)

# Loading the CLEANED AND FEATURED data from the previous step
DATA_PATH = '../data/processed/df_featured.csv'
try:
    df = pd.read_csv(DATA_PATH, parse_dates=['Date'])
    print("Featured data loaded successfully.")
    print(f"Data shape: {df.shape}")
except FileNotFoundError:
    print("ERROR: Please run Notebook 01 first to generate df_featured.csv")
```

```python
# Overall Sales Analysis
print("\n--- Starting EDA ---")

# 2.1 Distribution of Sales
plt.figure(figsize=(12, 6))
sns.histplot(df['Sales'], kde=True, bins=50)
plt.title('Distribution of Daily Sales per Store')
plt.show()

# 2.2 Overall Sales Trend Over Time
avg_sales_over_time = df.groupby('Date')['Sales'].mean()
plt.figure(figsize=(18, 7))
avg_sales_over_time.plot(linewidth=1.5, label='Daily Avg Sales')
avg_sales_over_time.rolling(window=30).mean().plot(linewidth=2.5, color='red', label='30-Day Rolling Avg')
plt.title('Average Daily Sales Across All Stores')
plt.legend()
plt.show()
```

```python
# Temporal Pattern Analysis

# 3.1 Sales by Month
plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x='Month', y='Sales', palette='viridis')
plt.title('Sales Distribution by Month')
plt.show()

# 3.2 Sales by Day of Week
plt.figure(figsize=(12, 7))
sns.boxplot(data=df, x='DayOfWeek', y='Sales', palette='rocket')
plt.title('Sales Distribution by Day of Week (1=Monday, 7=Sunday)')
plt.show()
```

```python
# Business Drivers Analysis

# Impact of Main Promotions (`Promo`)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Promo', y='Sales')
plt.title('Sales Comparison: With vs. Without Promotion')
plt.xlabel('Promotion Active (1=Yes, 0=No)')
plt.show()
```

```python

# Impact of Secondary Promotions (`IsPromoMonth`)
# This plot will now work because 'IsPromoMonth' was created in Notebook 1
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='IsPromoMonth', y='Sales')
plt.title('Sales Comparison: With vs. Without Secondary Promotion (Promo2)')
plt.xlabel('Secondary Promotion Active (1=Yes, 0=No)')
plt.show()

# Impact of Store Characteristics (StoreType and Assortment)
# This requires loading the original store.csv to get the character labels back for plotting
try:
    store_df_original = pd.read_csv('../data/raw/store.csv')
    df_viz = df.merge(store_df_original[['Store', 'StoreType', 'Assortment']], on='Store', how='left', suffixes=('', '_orig'))
    
    fig, ax = plt.subplots(1, 2, figsize=(18, 7))
    sns.boxplot(data=df_viz, x='StoreType_orig', y='Sales', ax=ax[0], order=['a', 'b', 'c', 'd'])
    ax[0].set_title('Sales by Store Type')
    sns.boxplot(data=df_viz, x='Assortment_orig', y='Sales', ax=ax[1], order=['a', 'b', 'c'])
    ax[1].set_title('Sales by Assortment Level')
    plt.tight_layout()
    plt.show()
except FileNotFoundError:
    print("Could not load original store.csv for visualization. Skipping StoreType/Assortment plots.")
```

```python

# Feature Correlation Analysis
features_to_correlate = [
    'Sales', 'Customers', 'Promo', 'SchoolHoliday', 'DayOfWeek', 'Month', 'WeekOfYear', 
    'CompetitionDistance', 'IsPromoMonth'
]

correlation_matrix = df[features_to_correlate].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Key Features')
plt.show()

print("\n--- Notebook 2 Complete ---")
```
